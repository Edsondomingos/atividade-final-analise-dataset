```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nomes = pd.read_table('nomes.csv', sep=',')
```

```python
nomes
```

```python
#retorna quantos dados nulos existem por coluna
nomes.isnull().sum()
```

```python
#retornando apenas as linhas que possuem alguma coluna com dado nulo
nomes[nomes.isnull().T.any()]
```

```python
#Retornar linhas que possuem dados nulo
nomes[nomes.alternative_names.isnull()]
```

```python
#Substitui apenas os valores nulos de nome por 'Sem alternativa'
nomes.loc[nomes.alternative_names.isnull(),'alternative_names'] = 'Sem alternativa'
```

```python
#Conferindo se os nulos foram trocados por 'Sem alternativa'
nomes[nomes.alternative_names == 'Sem alternativa']
```

```python
#Retornar linhas que possuem dados nulo
nomes[nomes.frequency_female.isnull()]
```

```python
#Substitui apenas os valores nulos de nome por '0'
nomes.loc[nomes.frequency_female.isnull(),'frequency_female'] = 0
```

```python
#Conferindo se os nulos foram trocados por '0'
nomes[nomes.frequency_female == 0]
```

```python
#Retornar linhas que possuem dados nulo
nomes[nomes.frequency_male.isnull()]
```

```python
#Substitui apenas os valores nulos de nome por '0'
nomes.loc[nomes.frequency_male.isnull(),'frequency_male'] = 0
```

```python
#Conferindo se os nulos foram trocados por '0'
nomes[nomes.frequency_male == 0]
```

```python
nomes.loc[:,['first_name','frequency_female', 'classification', 'alternative_names']].sort_values('frequency_female', ascending=False).head(30)
```
