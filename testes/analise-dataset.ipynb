## Quantidade de Canais para cada categoria

```python
qtd_canais_categoria = ytr.groupby(['category'])['category'].count().sort_values(ascending=True)
plt.bar(qtd_canais_categoria.index, qtd_canais_categoria.values)
plt.xticks(rotation=80)
plt.title('Quantidade de Canais por categorias')
plt.xlabel('Categorias')
plt.ylabel('Quantidade')
plt.grid(True)
plt.show()
```
