```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
copa_2014 = pd.read_csv('FIFA - 2014.csv')
```

```python
copa_2014
```

```python
jogos_jogados = np.array(copa_2014.loc[0:,'Games Played'])
qtd_jogos = np.sum(jogos_jogados)
print('Teve o total de',qtd_jogos,'jogos.')
```

```python
gol_contra = np.array(copa_2014.loc[0:,'Goals Against'])
qtd_gol_contra = np.sum(gol_contra)
print('Teve o total de',qtd_gol_contra,'gols contra.')
```
